### Test FRED Dataset

```python
import datetime as dt
import pandas as pd

from gs_quant.data import Dataset
from gs_quant.api.fred import FredDataApi
from IPython.display import display

pd.set_option('display.max_rows', 1000000)

FRED_API_KEY = <YOUR API KEY>
fredAPI = FredDataApi(api_key=FRED_API_KEY)
GDP_data = Dataset("GDP", fredAPI)
```

#### Get Data

```python
GDP_series = GDP_data.get_data()
GDP_series.head(20)
```

### Get Data

#### Get Data w/ Start Date

```python
GDP_data.get_data(start=dt.date(2000, 1, 1)).head()
```

#### Get Data w/ End Date

```python
GDP_data.get_data(end=dt.date(2000, 1, 1)).tail()
```

#### Get Data w/ Start and End Date

```python
GDP_series = GDP_data.get_data(start=dt.date(1970, 1, 1), end=dt.date(2000, 1, 1))

print("-----------------")
display(GDP_series.head())
print("-----------------")
display(GDP_series.tail())
```

#### Get Data as_of

```python
GDP_data.get_data(as_of=dt.date(2000, 1, 1)).tail(20)
```

#### Get Data since
